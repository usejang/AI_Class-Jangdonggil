```python
import pandas as pd
df_wine = pd.read_csv("./dataset/wine.csv", header=None)
df = df_wine.sample(frac=0.7)
df.info()
```

```python
import numpy as np
dataset = df.values
X = dataset[:, :-1].astype(float)
Y = dataset[:, -1]
```

```python
from keras.models import Sequential, load_model
from keras.layers.core import Dense
from keras.callbacks import ModelCheckpoint, EarlyStopping

import tensorflow as tf
import matplotlib.pyplot as plt
seed = 0
np.random.seed(seed)
tf.random.set_seed(3)
```

```python
#모델 설정
input_dim = X.shape[1]
model = Sequential()
model.add(Dense(30, input_dim=input_dim, activation='relu'))
model.add(Dense(12, activation='relu'))
model.add(Dense(8, activation='relu'))
model.add(Dense(1, activation='sigmoid'))

#모델 컴파일
model.compile(loss='binary_crossentropy',
             optimizer='adam',
             metrics=['accuracy'])
#checkpoint
checkpointer = ModelCheckpoint(filepath=modelpath, monitor='val_loss',
                              verbose=1, save_best_only=True)
model.fit(X, Y, validation_split=0.2, epochs=200, batch_size=200,
         callbacks=[checkpointer], verbose=0)

#모델 실행
model.fit(X, Y, epochs=200, batch_size=200)

#실행 결과 출력
print("\n Accuracy : %.4f" % (model.evaluate(X,Y)[1]))

model.save("wine.hd5")
```

```python
import os

MODEL_DIR = './model'
if not os.path.exists(MODEL_DIR):
    os.mkdir(MODEL_DIR)
    
modelpath = "./model/{epoch:02d}-{val_loss:.4f}.hdf5"
```
