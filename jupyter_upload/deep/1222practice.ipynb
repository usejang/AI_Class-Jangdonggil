```python
import sys
import cv2
import numpy as np

```

```python
cap = cv2.VideoCapture(0)

if not cap.isOpend():
    sys.exit('video open failed')
    cap.release()
    sys.exit()

w = cap.get(cv2.CAP_PROP_FRAME_WIDTH)
h= cap.get(cv2.CAP_PROP_FRAME_HEIGHT)
fps = cap.get(cv2.CAP_PROP_FPS*0.7)
fourcc = cv2.VideoWriter_fourcc(*'DIVX')

out = cv2.VideoWriter('out.avi', fourcc, fps, (w, h))


while True:
    ret, frame = cap.read()
    
    if not ret:
        print('video read failed')
        break
        
    edge = cv2.Canny(frame, 50, 150)
    out.write(frame)
    cv2.imshow(frame)
    cv2.imshow(edge)
    
        
    cv2.imshow(frame)
    if cv2.waitKey(30) == 27:
        break
        
cap.release()
cv2.destroAllWindows()
```
