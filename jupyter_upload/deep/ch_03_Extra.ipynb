```python
import numpy as np
import cv2
import sys


oldx = oldy = -1

def call_mouse(event, x, y, flags, param):
    global oldx, oldy
    if event == cv2.EVENT_LBUTTONDOWN:
        oldx, oldy = x, y
    
    elif event == cv2.EVENT_MOUSEMOVE:
        if flags == cv2. EVENT_FLAG_LBUTTON:
            cv2.line(img, (oldx, oldy), (x, y),
                    (0, 0, 255), 3, cv2.LINE_AA)
            cv2.imshow('image', img)
            oldx, oldy = x, y
        
img = np.ones((400, 400, 3), np.uint8)*255

cv2.namedWindow('image')
cv2.setMouseCallback('image', call_mouse, img)

cv2.imshow('image', img)
cv2.waitKey
cv2.destroyAllWindows
```

```python
import numpy as np
import cv2
import sys
import matplotlib.pyplot as plt

src1 = np.zeros((256, 256), np.uint8)
cv2.rectangle(src1, (10, 10), (127, 248), 255, -1)

src2 = np.zeros((256, 256), np.uint8)
cv2.circle(src2, (128, 128), 100, 255, -1)

# cv2.imshow('src1', src)
# cv2.imshow('src2', src)
dst_and = cv2.bitwise_and(src1, src2)
dst_or = cv2.bitwise_or(src1, src2)
dst_xor = cv2.bitwise_xor(src1, src2)
dst_not = cv2.bitwise_not(src1)
dst_test = cv2.bitwise_xor(src1, dst_or)
plt.figure(figsize = (16, 6))
plt.subplot(241), plt.imshow(src1, 'gray'), plt.axis('off')
plt.subplot(242), plt.imshow(src2, 'gray'), plt.axis('off')
plt.subplot(243), plt.imshow(dst_and, 'gray'), plt.axis('off'), plt.title('and')
plt.subplot(244), plt.imshow(dst_or, 'gray'), plt.axis('off'), plt.title('or')
plt.subplot(245), plt.imshow(dst_xor, 'gray'), plt.axis('off'), plt.title('xor')
plt.subplot(246), plt.imshow(dst_not, 'gray'), plt.axis('off'), plt.title('not')
plt.subplot(247), plt.imshow(dst_test, 'gray'), plt.axis('off'), plt.title('test')
# cv2.imshow('dst', dst)

cv2.waitKey()
cv2.destroyAllWindows()
```
