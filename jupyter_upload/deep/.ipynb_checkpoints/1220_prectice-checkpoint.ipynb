```python
src = cv2.imread('fig/airplane.bmp',1)
mask = cv2.imread('fig/mask_plane.bmp',0)
dst = cv2.imread('fig/field.bmp',1)

if src is None or mask is None or dst is None:
    print('Image read failed!')
    sys.exit()
    
dst = cv2.copyTo(src, mask)

cv2.imshow('src', src)
cv2.imshow('mask', mask)
cv2.imshow('dst', dst)

cv2.waitKey()
cv2.destroAllWindows()
```

```python
img1 = cv2.imread('fig/cow.png')
img2 = cv2.imread('fig/green.png')
img1_gray = cv2.cvtColor(img1, cv2.COLOR_BGR2GRAY)

h, w = img1.shape[:2]
img2 = cv2.resize(img2, (w, h), cv2.INTER_AREA)
retval, mask = cv2.threshold(img1_gray, 240, 255, cv2. THRESH_BINARY_INV)

cv2.copyTo(img1, mask, img2)

cv2.imshow('img1', img1)
cv2.imshow('img2', img2)
cv2.imshow('mask', mask)

cv2.waitKey()
cv2.destroyAllWindows()
```

```python
cap = cv2.VideoCapture(0)
if not cap.isOpened():
    print('video open failed')
    sys.exit()

w = round(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
h = round(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fps = round(cap.get(cv2.CAP_PROP_FRAME_FPS)*0.7)
fourcc = cv2.VideoWriter_fourcc(*'DIVX')

out = cv2.VideoWriter('output.avi', fourcc, fps, (w, h))

while True:
    ret, frame = cap.read()
    
    if not ret:
        print('video read failed')
        break
        
    edge = cv2.Canny(frame, 50, 150)
    out.write(frame)
    cv2.imshow('frame', frame)
    cv2.imshow('edge', edge)
    if cv2.waitKey(30) == 27:
        break
        
cap.release()
cv2.destroyAllWindows()
```

```python
img = cv2.imread('fig/flower.jpg', cv2.IMREAD_GRAYSCALE)

cv2.imshow('image', img)

while True:
    key = cv2.waitKey()
    
    if key == 27:
        break
    elif key == ord('i'):
        img = ~ img
        cv2.imshow('image', img)
    elif key == ord('e'):
        img = cv2.Canny(img, 50, 150)
        cv2.imshow('image', img)

cv2.waitKey()
cv2.destroyAllWindows()
```

```python
import numpy as np
import cv2
import sys

def call_mouse(event, x, y, flags, pram):
    global oldx, oldy
    
    if event == cv2.EVENT_LBUTTONDOWN:
        oldx, oldy = x, y
        
    elif event == cv2.EVENT_MOUSEMOVE:
        if flags == cv2.EVENT_FLAG_LBUTTON:
            cv2.line(img, (oldx, oldy), (x, y), (255, 0, 255), 4, cv2.LINE_AA)
            cv2.imshow('image', img)
            oldx, oldy = x, y
#             print(x, y)



img = np.ones((480, 480, 3), np.uint8)*255
cv2.namedWindow('image')
cv2.setMouseCallback('image', call_mouse, img)

cv2.imshow('image', img)

while True:
    key = cv2.waitKey()
    if key == 27:
        break
        
    elif key == ord('s'):
        cv2.imwrite('signature.png', img)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
import numpy as np
import cv2
import sys
import matplotlib.pyplot as plt

src1 = cv2.imread('fig/lenna256.bmp', 0)
# src2 = np.zeros_like('fig/lenna256.bmp', np.uint8)
src2 = np.zeros((256, 256), np.uint8)
cv2.circle(src2, (128, 128), 100, 200, -1)
cv2.circle(src2, (128, 128), 50, 50, -1)
# print(src2.shape)

dst1 = cv2.add(src1, src2)
dst2 = cv2.addWeighted(src1, 0.5, src2, 0.5, 0.0)
dst3 = cv2.subtract(src1, src2)
dst4 = cv2.absdiff(src1, src2)

# cv2.imshow('src1', src1)
# cv2.imshow('src2', src2)
# cv2.imshow('dst1', dst1)
# cv2.imshow('dst2', dst2)
# cv2.imshow('dst3', dst3)
# cv2.imshow('dst4', dst4)

plt.figure(figsize = (12, 6))
plt.subplot(231), plt.imshow(src1, 'gray'), plt.axis('off')
plt.subplot(232), plt.imshow(src2, 'gray'), plt.axis('off')
plt.subplot(233), plt.imshow(dst1, 'gray'), plt.axis('off')
plt.subplot(234), plt.imshow(dst2, 'gray'), plt.axis('off')
plt.subplot(235), plt.imshow(dst3, 'gray'), plt.axis('off')
plt.subplot(236), plt.imshow(dst4, 'gray'), plt.axis('off')


cv2.waitKey()
cv2.destroyAllWindows()
```

```python
import numpy as np
import cv2
import sys
import matplotlib.pyplot as plt

src1 = np.zeros((256, 256), np.uint8)
cv2.rectangle(src1, (10, 10), (127, 248), 255, -1)

src2 = np.zeros((256, 256), np.uint8)
cv2.circle(src2, (128, 128), 100, 255, -1)

# cv2.imshow('src1', src)
# cv2.imshow('src2', src)
dst_and = cv2.bitwise_and(src1, src2)
dst_or = cv2.bitwise_or(src1, src2)
dst_xor = cv2.bitwise_xor(src1, src2)
dst_not = cv2.bitwise_not(src1)
dst_test = cv2.bitwise_xor(src1, dst_or)
plt.figure(figsize = (16, 6))
plt.subplot(241), plt.imshow(src1, 'gray'), plt.axis('off')
plt.subplot(242), plt.imshow(src2, 'gray'), plt.axis('off')
plt.subplot(243), plt.imshow(dst_and, 'gray'), plt.axis('off'), plt.title('and')
plt.subplot(244), plt.imshow(dst_or, 'gray'), plt.axis('off'), plt.title('or')
plt.subplot(245), plt.imshow(dst_xor, 'gray'), plt.axis('off'), plt.title('xor')
plt.subplot(246), plt.imshow(dst_not, 'gray'), plt.axis('off'), plt.title('not')
plt.subplot(247), plt.imshow(dst_test, 'gray'), plt.axis('off'), plt.title('test')
# cv2.imshow('dst', dst)

cv2.waitKey()
cv2.destroyAllWindows()
```
