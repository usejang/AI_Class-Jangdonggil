```python
src = cv2.imread('fig/Hawkes.jpg', 0)
# smin, smax,min_loc, max_loc = cv2.minMaxLoc(src)
# print(smin, smax)
# print(min_loc, smax)
# dst = 255*(src-smin)/(smax-smin) + (0, 0, 255).astype(unp.uint8)

dst_norm = cv2.normalize(src, None, 0, 255, cv2.NORM_MINMAX, -1)
dst_equal = cv2.equalizeHist(src)

hist_src =cv2. calcHist([src], [0], None, [256], [0, 256])
hist_equal =cv2. calcHist([dst_equal], [0], None, [256], [0, 256])
hist_norm =cv2. calcHist([dst_norm], [0], None, [256], [0, 256])

plt.plot(hist_src, color = "b")
plt.plot(hist_norm, color = 'g')
plt.plot(hist_equal, color = 'r')
plt.show()

cv2.imshow('src', src)
cv2.imshow('dst_norm', dst_norm)
cv2.imshow('dst_equal', dst_equal)
cv2.waitKey()
cv2.destroyAllWindows
```

```python
import numpy as np
import cv2
import sys
import matplotlib.pyplot as plt


src = cv2.imread('fig/pexels-mike-145685.jpg', 0)

h, w = src.shape[:2]

src = cv2.resize(src, (w//6, h//6), cv2.INTER_AREA)
src1 = src.copy()
src = cv2.addWeighted(src, 0.1, src1, 0, 150)

dst_norm = cv2.normalize(src, None, 0, 255, cv2.NORM_MINMAX, -1)
dst_equal = cv2.equalizeHist(src)

hist_src = cv2.calcHist([src], [0], None, [256], [0, 256])
hist_equal = cv2.calcHist([dst_equal], [0], None, [256], [0, 256])
hist_norm = cv2.calcHist([dst_norm], [0], None, [256], [0, 256])

plt.plot(hist_src, color = "b")
plt.plot(hist_norm, color = 'g')
plt.plot(hist_equal, color = 'r')
plt.show()

cv2.imshow('src', src)
cv2.imshow('dst_norm', dst_norm)
cv2.imshow('dst_equal', dst_equal)
cv2.waitKey()
cv2.destroyAllWindows
```
