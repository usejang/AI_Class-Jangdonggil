```python
#개별 원소에 함수 매핑 : 시리즈객체,apply(함수명)

df = titanic.loc[ :, ['age', 'fare']]
df.head()
```

```python
def add_10(n):
    return n+10
def add_two_10(a,b):
    return a+b

sr1 = df['age'].apply(add_10)    #(lambda x: x+10)
sr1.head()
sr2 = df['age'].apply(add_two_10, b=20)
sr2.head()
```

```python
#개별 원소에 함수 매핑 : 데이터프레임, applymap(함수명)
df.head()
df.applymap(add_10)    #df.applymap(lambda x: x+ 10)
```

```python
#데이타 프레임의 각 열에 함수 매핑
def missing_value(ser):
    return ser.isnull()

result = df.apply(missing_value, axis=0)
result.head()
```

```python
def max_min(x):
    return x.max() - x.min()
result = df.apply(max_min, axis=0)
result.head()
```

```python
def max_min(x):
    return x.max() - x.min()
result = df.apply(max_min, axis=1)
result.head()
```

```python
def max_min(x):
    return x.max() - x.min()

result = df.apply(lambda x: max_min, axis=1)
result.head()
df['sum'] = df.apply(lambda x: x.sum(), axis=1)
df.head()
```

```python
#시리즈 함수 매핑 : 시리즈.apply(함수명)
#데이터프레임에 함수 매핑 : 데이터프레임.applymap(함수명)
#데이터 프레임 각 열에 함수 매핑 : 데이터프레임.apply(함수명) axis=0 default
#데이터 프레임 각 행에 함수 매핑 : 데이터프레임.apply(함수명,  axis=1)
#데이터 프레임 함수매핑 : 데이터프레임.pip(함수명) >> 반환하는 종류에 따라 반환하는 타입 정해짐
def missing_value(x):
    return x.isnull()
def missing_count(x):
    return missing_value(x).sum()
def total_number_missing(x):
    return missing_count(x).sum()

result_df = df.pipe(missing_value)   #false,True
result_ser = df.pipe(missing_count)
result_val = df.pipe(total_number_missing)

print(result_df.head())
print(result_ser.head())
print(result_val)
```

```python
df = titanic.loc[ :, ['age','class','fare','survived','sex']]

grouped = df.groupby('class')
age_mean = grouped.age.mean()
age_std = grouped.age.std()
age_mean

for key, group in grouped.age:        #(승객의 (나이-평균) / 표준편차)
    group_zeros = (group - age_mean.loc[key])/age_std.loc[key]
    print("key : ",key)
    print(group_zeros.head(3))
    
def z_score(x):
    return (x - x.mean())/ x.std()

age_zeros = grouped.age.transform(z_score)   #각 데이터에 함수를 적용
age_zeros.loc[[1,9,0]]
                                                            
```
